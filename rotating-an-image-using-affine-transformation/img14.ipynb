```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('57515.jpg')
```

```python
plt.imshow(img[...,::-1])
```

```python
rot_deg = -10
theta = rot_deg*np.pi/180
Tx = 10
Ty = 0
```

```python
height,width,_ = img.shape
```

```python
T = np.float32([[np.cos(theta), -np.sin(theta), Tx],[np.sin(theta), np.cos(theta), Ty]])
```

```python
T
```

```python
img_translation = cv2.warpAffine(img, T, (width,height))
```

```python
plt.imshow(img_translation[...,::-1])
```
