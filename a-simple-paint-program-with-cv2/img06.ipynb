```python
import numpy as np
import argparse
import cv2
import matplotlib.pyplot as plt
```

```python
img = np.zeros((512,512,3))
drawing = False
ix = 0
iy = 0
def draw(event,x,y,flags,param):
    global ix, iy, drawing
    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        ix = x
        iy = y
    if event == cv2.EVENT_MOUSEMOVE:
        if drawing == True:
            cv2.line(img, pt1 = (ix,iy) ,pt2 = (x,y) , color = (255,255,255), thickness = 3)
            ix = x
            iy = y
    if event == cv2.EVENT_LBUTTONUP:
        drawing = False
        
        
cv2.namedWindow('Window')
cv2.setMouseCallback('Window', draw)

while True:
    cv2.imshow('Window', img)
    if cv2.waitKey(1) & 0xFF == 27:
        break
    
cv2.destroyAllWindows()
plt.imshow(img[...,::-1])
```

```python
img = np.zeros((512,512,3))
cv2.namedWindow('img')
drawing = False
ix = 0
iy = 0
def nothing(x):
    pass
cv2.createTrackbar('R','img',0,255,nothing)
cv2.createTrackbar('G','img',0,255,nothing)
cv2.createTrackbar('B','img',0,255,nothing)
cv2.createTrackbar('Thickness','img',0,100,nothing)
def draw(event,x,y,flags,param):
    global ix, iy, drawing
    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        ix = x
        iy = y
    if event == cv2.EVENT_MOUSEMOVE:
        if drawing == True:
            r = cv2.getTrackbarPos('R','img')
            g = cv2.getTrackbarPos('G','img')
            b = cv2.getTrackbarPos('B','img')
            thick = cv2.getTrackbarPos('Thickness','img')
            cv2.line(img, pt1 = (ix,iy) ,pt2 = (x,y) , color = (r,g,b), thickness = thick)
            ix = x
            iy = y
    if event == cv2.EVENT_LBUTTONUP:
        drawing = False
        
        
cv2.namedWindow('img')
cv2.setMouseCallback('img', draw)

while True:
    cv2.imshow('img', img)
    if cv2.waitKey(1) & 0xFF == 27:
        break
    
cv2.destroyAllWindows()
plt.imshow(img[...,::-1])
```
